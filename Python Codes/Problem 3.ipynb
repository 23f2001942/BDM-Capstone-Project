```python
months = ['October', 'November', 'December']
revenues = [44860, 47932, 48109]
outsourced_costs = [37000, 30000, 39000]
inhouse_cost = 22500
discount_rate = 0.008  # monthly (10% annually)

net_benefits = [rev - inhouse_cost for rev in revenues]
npv = sum(benefit / ((1 + discount_rate) ** (i + 1)) for i, benefit in enumerate(net_benefits))

import pandas as pd
df = pd.DataFrame({
    'Month': months,
    'Revenue': revenues,
    'Outsourced Cost': outsourced_costs,
    'In-House Cost': [inhouse_cost]*3,
    'Net Benefit': net_benefits
})

print(df)
print(f"\nEstimated NPV over 3 months at 10% annual discount rate: ₹{npv:.2f}")
```
